```python
import math

class Shape:
    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    @property
    def length(self):
        return self.__length

    @length.setter
    def length(self, value):
        if value >= 0:
            self.__length = value
        else:
            print("Error: Length must be non-negative.")

    @property
    def width(self):
        return self.__width

    @width.setter
    def width(self, value):
        if value >= 0:
            self.__width = value
        else:
            print("Error: Width must be non-negative.")

    def calculate_area(self):
        pass  # To be overridden by subclasses

class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        super().__init__(radius, None)  # Pass None as width for Circle
        self.__radius = radius

    @property
    def radius(self):
        return self.__radius

    @radius.setter
    def radius(self, value):
        if value >= 0:
            self.__radius = value
        else:
            print("Error: Radius must be non-negative.")

    def calculate_area(self):
        return math.pi * (self.radius ** 2)
class Triangle(Shape):
    def calculate_area(self):
        # Area of a triangle: (base * height) / 2
        return (self.length * self.width) / 2

# Create instances of Rectangle and Circle
rectangle = Rectangle(9, 8)
circle = Circle(9)
triangle = Triangle(5, 4)
# Calculate and print areas
print(rectangle.calculate_area())
print(circle.calculate_area())
print(triangle.calculate_area())

print("Demonstrate polymorphism by creating a list of different shapes")
shapes = [rectangle, circle, triangle]

# Calculate and print areas using a loop
for shape in shapes:
    print(shape.calculate_area())

# # Use the setters for Shape's length and width
# rectangle.length = 12
# rectangle.width = 6

# # Recalculate and print the area
# print(rectangle.calculate_area())
```
